```python
# Game : 7 to smoke 
# Rule : 1-1 battle. the 1st person who wins 7 times will win the whole game . ( doesn't have to be 7 in a row)
# People :
    # Judge : Natalie
    # Battler : 1. boywonder 2. jahsan 3. shez 4. rashaad 5. future 6. tino
# Criteria : 
    # 1. Foundation & Technique 2. Musicality  3. Dopeness 4. Space 5. Performance 6. Creativity 7. Variations of styles

# Pseudo:
# 1. Class - People 
    # create Judge : 1 (Random)
        # use super
    # create 6 Battler 
        # use super
# 2. Class - Game
    # 1. create each criteria
    # 2. creat who win 
    # 3. create keep track of each person # of win

    # Play game(): - this game will loop loop so use while true
    # 0. intially no winner , the 1st round
    # 1. Randomly pick "1 person from Battler" to battle  "the other 1 Battler" 
    # 2. score by Judege for Totals score of Criteria (random). ( print: "Person 1 vs Person 2 " , the score of each Criteria is : 
    # 3. decide who win (highest total scores) . (print Natalie voted for ___)
    # 4. Save: winner (highest total scores) + 1 . winner keep going - show on output (already win x rounds) too. 
    # 5. Randomly pick "another Battler" to battle the "winner" 
    # 6. repeat 
    # 7. break when 1 winner got 7 rounds. 

# 3. instance = judge, battler
# 4. play game()
```

```python
import random

# people start
class Person:
    def __init__(self, name):
        self.name = name
        self.wins = 0

class Judge(Person):
    def __init__(self, name):
        super().__init__(name)

class Battler(Person):
    def __init__(self, name):
        super().__init__(name)
# people end 

# game start
class Game:
    def __init__(self, judge, battlers):
        self.judge = judge
        self.battlers = battlers
        self.criteria = [
            "Foundation & Technique",
            "Musicality",
            "Dopeness",
            "Space",
            "Performance",
            "Creativity",
            "Variations of Styles"
        ]

    #play the game
    #1. pick battler
    #2. score
    #3. who has higher score = winner
    #4. update the winner
    #5. break if winner =7
    def play_game(self):
        
       
        winner = None # initially - no winner at the 1st round
        
        #this game keep goin in an infinite loop using while True
        while True:
            
        # 1. pick battler start
            if not winner: #1.1 if no winner - random
                battler1 = random.choice(self.battlers) #1.1.1 random from the self.battlers from Game from people
            else:
                battler1 = winner #1.2 battler1 = winner
            battler2 = random.choice(self.battlers)
            
            # 1.1 to avoid battler 1 & 2 are the same
            while battler1 == battler2: #
                battler2 = random.choice(self.battlers) # battler2 choose another one
                
            
            # now we picked, then print who they are
            print(f"{battler1} vs {battler2}")
            
        # pick battler end
            
            
        # 2. Score

            # 2.1. each criteria score
            scores_battler1 = [random.randint(1, 10) for _ in range(7)]
            scores_battler2 = [random.randint(1, 10) for _ in range(7)]

            ### print the details of each criteria
            print("Scores:")
            
            # print all the criteria & score
            for i in range(7):
                print(f"  {self.criteria[i]}: {scores_battler1[i]} vs {scores_battler2[i]}")
                            # this criteria : batter1 score v.s. battler2 score ---- keep printing for 7 times
        
             # 2.2 sum total score
            total_score_battler1 = sum(scores_battler1)
            total_score_battler2 = sum(scores_battler2)
               
        # score end
        
        # 3. pick who has higher total score = the winner 
            if total_score_battler1 > total_score_battler2:
                winner = battler1
            elif total_score_battler1 < total_score_battler2:
                winner = battler2
            else:
                winner = random.choice([battler1, battler2]) # if they have the same score, since they are old, they can't battle again, just pick either one lol
            # well , if you want them keep going, then:
            # else:
                # continue # maybe print : tie! another round!
                
            print(f" {battler1} total score is {total_score_battler1} & {battler2} {total_score_battler2}")
            
            print(f"{self.judge} voted for {winner}")

        # 4. update winner count
            winner.wins += 1

            print(f"{winner} wins {winner.wins} rounds so far!\n")
            
            
        #5. if winner > 7
            if winner.wins >= 7:
                print(f"Congratulations! {winner} wins the 7 to smoke!")
                break


judge = Judge("Natalie")

battlers = [
    Battler("boywonder"),
    Battler("jahsan"),
    Battler("shez"),
    Battler("rashaad"),
    Battler("future"),
    Battler("tino")
]

game = Game(judge, battlers)

game.play_game()
```

```python
import random

# People start
class Person:
    def __init__(self, name):
        self.name = name
        self.wins = 0

class Judge(Person):
    def __init__(self, name):
        super().__init__(name)

class Battler(Person):
    def __init__(self, name):
        super().__init__(name)
# People end 

# Game start
class Game:
    def __init__(self, judge, battlers):
        self.judge = judge
        self.battlers = battlers
        self.criteria = [
            "Foundation & Technique",
            "Musicality",
            "Dopeness",
            "Space",
            "Performance",
            "Creativity",
            "Variations of Styles"
        ]
        self.current_battler1_index = 0

    def get_next_battler2(self):
        self.current_battler1_index = (self.current_battler1_index + 1) % len(self.battlers)
        return self.battlers[self.current_battler1_index]

    def play_game(self):
        winner = None
        battler1 = None
        
        while True:
            # Select battler 1 for the first round or use the winner from the previous round
            if battler1 is None:
                battler1 = self.battlers[self.current_battler1_index]
            else:
                battler1 = winner

            # Get the next battler 2 from the list
            battler2 = self.get_next_battler2()
            
            # Print the battlers for this round
            print(f"{battler1.name} vs {battler2.name}")
            
            # Scoring
            scores_battler1 = [random.randint(1, 10) for _ in range(7)]
            scores_battler2 = [random.randint(1, 10) for _ in range(7)]

            # Print scores for each criterion
            print("Scores:")
            for i in range(7):
                print(f"  {self.criteria[i]}: {scores_battler1[i]} vs {scores_battler2[i]}")
            
            # Total score calculation
            total_score_battler1 = sum(scores_battler1)
            total_score_battler2 = sum(scores_battler2)
            
            # Winner selection
            if total_score_battler1 > total_score_battler2:
                winner = battler1
            elif total_score_battler1 < total_score_battler2:
                winner = battler2
            else:
                continue
            
            print(f"Total score: {battler1.name}: {total_score_battler1}, {battler2.name}: {total_score_battler2}")
            print(f"{self.judge.name} voted for {winner.name}")

            # Update winner count
            winner.wins += 1
            print(f"{winner.name} wins {winner.wins} rounds so far!\n")

            # Check if any battler has won 7 rounds
            if winner.wins >= 7:
                print(f"Congratulations! {winner.name} wins the 7 to smoke!")
                break

judge = Judge("Natalie")

battlers = [
    Battler("jahsan"),
    Battler("future"),
    Battler("rashaad"),
    Battler("shez"),
    Battler("boywonder"),
    Battler("tino")
]

game = Game(judge, battlers)
game.play_game()
```
